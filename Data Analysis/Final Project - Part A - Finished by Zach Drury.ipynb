```python
# 3-4) Show the death rate in MY COUNTRY , 
# Is it above or lower than the worldwide death rate?
# (3 pts)

number = random.randint(0,10000)
print("Zach Drury , dr21637 ", number )
print('------------------------------')

res = df[df['location']== 'United Arab Emirates']
tot_new_cases = res['new_cases'].sum()

res = df[df['location'] == 'United Arab Emirates']
tot_new_deaths = res['new_deaths'].sum()

rate1 = tot_new_deaths / tot_new_cases
print('Show the death rate in United Arab Emirates:',format(rate1, '.2%'))
      
rate2 = df['new_deaths'].sum()/df['new_cases'].sum()
print('Show the death rate in the World:',format(rate2, '.2%'))
print()
if rate1 > rate2:
    print('United Arab Emirates above the world level.')
else:
    print('United Arab Emirates below the world level.')
```

```python
# 5-6) Show the average new_cases_per_million people in Brazil 
# Is it above or lower than the worldwide level.
# (6 pts)
number = random.randint(0,10000)
print("Zach Drury , dr21637 ", number )
print('------------------------------')

res = df[df['location'] =='United Arab Emirates']
num1 = res['new_cases_per_million'].mean()
print('Show the average new cases per million in the United Arab Emirates:', format(num1, '.0f'))

num2 = df['new_cases_per_million'].mean()
print('Show the average new cases per million in the World:', format(num2, '.0f'))
print()
if num1 > num2:
    print("United Arab Emirates is above the world level.")
else:
    print("United Arab Emirates is below the world level.")
```
