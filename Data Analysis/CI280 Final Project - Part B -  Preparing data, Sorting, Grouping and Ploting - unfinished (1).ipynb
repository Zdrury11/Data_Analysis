```python
# 1)  (3 pts) Find the top 10 countries which have the most covid cases
number = random.randint(0,10000)
print("Zach Drury , dr21637 ", number )
print('------------------------------')


# Group all cases by location , become series
cases = df.groupby(df.location).new_cases.sum()
cases = cases.sort_values(ascending = False)
cases

res = cases.head(10)
countries = res.index

print('10 countries which have the most covid cases')
print('Country                 Covid cases')
print('--------------------------------------------')
for country in countries:
    print('{0: <20}'.format(country), '\t', format(res[country], '.0f'))
```

```python
# 2) Use bar chart to show the top 10 countries which have the most covid cases. (3 pts)
number = random.randint(0,10000)
print("Zach Drury , dr21637 ", number )
print('------------------------------')

res = cases.head(10)
res
res = cases.head(10)
print(res.index)
print(res.values)
res.plot(kind = 'bar', title= '10 countries which have the most Covid cases')
```

```python
# 3) Show worldwide, which ten day has the most new cases? (3pts)
#    Use line chart to show the top 10 days which have the most new covid cases        
#    Draw bar chart
number = random.randint(0,10000)
print("Zach Drury , dr21637 ", number )
print('------------------------------')

cases = df.groupby(df.date).new_cases.sum()
cases = cases.sort_values(ascending = False)

res = cases.head(10)
res = res.sort_index()
days = res.index

print('Ten days which have the most Covid cases')
print('Date                        Covid Cases')
print('---------------------------------------')
for day in days:
    print(day, '\t', format(res[day], '.0f'))
    
plt.title('Top ten days which have the most new Covid cases')
plt.xlabel("Date")
plt.ylabel("Numbers")

plt.plot(res.index,res.values, color='b')
res.plot(kind ='line', rot=45, linestyle='dashed', marker='o')
```

```python
# (3  pts)
# 4) On 2020-11-12, which ten countries has the most new cases? 
#    Use bar chart to show the top 10 location which have the most new covid cases 
number = random.randint(0,10000)
print("Zach Drury , dr21637 ", number )
print('------------------------------')

res = df[df.date == '2020-11-12']

cases = res.sort_values(by='new_cases', ascending = False)

res = cases.head(10)[['location','new_cases']]
res = res.sort_index()
print(res)

print('On 2020-11-12')
print('10 countries which have the most Covid cases on 2020-11-12')
print('Country                                 Covid Cases')
print('-----------------------------------------------------')
for i in range(len(res)):
    country = res['location'].iloc[i]
    new_cases = res['new_cases'].iloc[i]
    print('{0: <20}'.format(country), '\t', format(new_cases, '0f'))
    
#    Draw bar chart
# Add title, xlabel, ylabel
plt.title('10 countries which have the most Covid cases on 2020-11-12')
plt.xlabel('Date')
plt.xlabel('Numbers')

plt.bar(res['location'],res['new_cases'], color='b')
```

```python
#  5)  Show last seven days, the new cases in Brazil/US/India/Russia (Chose one country),
#      Plot a line chart
number = random.randint(0,10000)
print("Zach Drury , dr21637 ", number )
print('------------------------------')

cases = df[df.location == 'United Arab Emirates'].sort_values(by = 'date', ascending = True)
cases
res = cases[['date','new_cases']].tail(7)

print('In the past seven days, the new cases in the United Arab Emirates')
print('Date                               Covid Cases')
print('--------------------------------------------')

dates = res['date']
for i in res.index:
    print(res['date'][i], '\t', format(res['new_cases'][i], '.0f'))


plt.title('New Covid cases in the past ten days in the United Arab Emirates')
plt.xlabel('Date')
plt.xlabel('Numbers')

plt.plot(res['date'],res['new_cases'], color='b')
```

```python
# 6-7) Group by contient, show total of new covid cases for each contient and sorted by total number
#    Plot the result by pie chart (3 pts)
number = random.randint(0,10000)
print("Zach Drury , dr21637 ", number )
print('------------------------------')

df
res = df.groupby(df.continent).new_cases.sum().sort_values(ascending = False)

print('Total new Covid cases by Continent')
print('--------------------------------------')

names = res.index
for name in names:
    print('{0: <20}'.format(name), '\t', format(res[name], '.0f'))

res.plot(kind='pie', title='Covid new cases of percentages by Continent',autopct= '%1.1f%%')
```
